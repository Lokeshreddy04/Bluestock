```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine( "postgresql://postgres:Lokesh$04@localhost:5432/fintech_dw")
```

```python


sector_df = pd.read_excel(
    '../data/sector_mapping.csv.xlsx'
)

dim_sector = (
    sector_df[["sector"]]
    .drop_duplicates()
)

dim_sector.rename(
    columns={
        "sector": "sector_name"
    },
    inplace=True
)

dim_sector["sector_code"] = (
    dim_sector["sector_name"]
    .str.upper()
    .str[:4]
)

dim_sector["description"] = (
    dim_sector["sector_name"]
)

print(dim_sector.head())
```

```python

dim_sector.to_sql(
    "dim_sector",
    engine,
    if_exists="append",
    index=False
)

print("✅ dim_sector loaded")
```

```python
analysis = pd.read_csv(
    "../data/clean/analysis_clean.csv"
)

analysis.rename(
    columns={
        "company_id": "symbol"
    },
    inplace=True
)

analysis.to_sql(
    "fact_analysis",
    engine,
    if_exists="append",
    index=False
)

print("✅ fact_analysis loaded")
```
