```python
import pandas as pd
```

```python
sector_df = pd.read_excel(
    '../data/sector_mapping.csv.xlsx')

print("Company columns:", dim_company.columns.tolist())
print("Sector columns:", sector_df.columns.tolist())
```

```python
# 1. Make sure dim_company is defined
dim_company = pd.read_csv('../data/clean/companies_clean.csv')

# 2. Merge using 'id' from the left dataframe and 'symbol' from the right dataframe
dim_company = dim_company.merge(
    sector_df, 
    left_on="id",       # The symbol column in dim_company
    right_on="symbol",  # The symbol column in sector_df
    how="left"
)
```

```python
# Clean trailing/leading spaces
dim_company['id'] = dim_company['id'].astype(str).str.strip()
sector_df['symbol'] = sector_df['symbol'].astype(str).str.strip()

# Retry the merge
dim_company_clean = dim_company.drop(columns=['symbol', 'sector_name'], errors='ignore') # drop old failed columns
dim_company_clean = dim_company_clean.merge(sector_df, left_on="id", right_on="symbol", how="left")
dim_company_clean.rename(columns={"sector": "sector_name"}, inplace=True)
print(dim_company_clean[['id', 'company_name', 'sector_name']].head())
```
