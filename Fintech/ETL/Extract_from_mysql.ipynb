```python
import pandas as pd
import numpy as np


# ---------- COMMON CLEANING ----------
def clean_common(df):
    df.replace(["NULL", "null", ""], np.nan, inplace=True)
    return df


def convert_numeric(df):
    for col in df.columns:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace(",", "", regex=False)
            .str.replace("%", "", regex=False)
        )
        df[col] = pd.to_numeric(df[col], errors="ignore")
    return df


# ---------- LOAD + FIX HEADER ----------
def load_with_header_fix(path, target_column):
    df_raw = pd.read_csv(path, header=None)

    header_row = None
    for i in range(10):
        row_values = df_raw.iloc[i].astype(str).str.lower().tolist()
        if target_column in row_values:
            header_row = i
            break

    if header_row is None:
        raise Exception(f"Header not found for {target_column}")

    df = pd.read_csv(path, skiprows=header_row)
    df.columns = df.columns.str.strip()

    return df


# ---------- CLEAN COMPANIES ----------
def clean_companies():
    df = load_with_header_fix("data/raw/companies.csv", "company_name")

    df = clean_common(df)

    df["company_name"] = df["company_name"].astype(str).str.strip()

    df.to_csv("data/clean/companies_clean.csv", index=False)

    print("✅ companies cleaned")


# ---------- CLEAN FINANCIAL ----------
def clean_financial(table):
    df = load_with_header_fix(f"data/raw/{table}.csv", "id")

    df = clean_common(df)
    df = convert_numeric(df)

    df.to_csv(f"data/clean/{table}_clean.csv", index=False)

    print(f"✅ {table} cleaned")


# ---------- MAIN ----------
def main():
    clean_companies()

    for table in ["balancesheet", "profitandloss", "cashflow"]:
        clean_financial(table)


main()
```
