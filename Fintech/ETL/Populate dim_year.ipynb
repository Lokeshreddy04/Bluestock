```python
import pandas as pd

years = []

for year in range(2013, 2025):

    years.append({
        "year_label": f"Mar {year}",
        "fiscal_year": year,
        "quarter": "Q4",
        "is_ttm": False,
        "is_half_year": False,
        "sort_order": year
    })

years.append({
    "year_label": "TTM",
    "fiscal_year": 2025,
    "quarter": "TTM",
    "is_ttm": True,
    "is_half_year": False,
    "sort_order": 9999
})
```

```python
dim_year = pd.DataFrame(years)

print(dim_year)
```

```python
from sqlalchemy import create_engine

engine = create_engine( "postgresql://postgres:Lokesh$04@localhost:5432/fintech_dw")

dim_year.to_sql(
    "dim_year",
    engine,
    if_exists="append",
    index=False
)

print("✅ dim_year loaded")
```
